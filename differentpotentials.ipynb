```python
import pandas as pd 
import numpy as np
import seaborn as sns
import csv
from matplotlib import pyplot as plt
```

```python
X = []
Y = []
with open('res1.txt', 'r') as datafile:
    plotting = csv.reader(datafile)
    for rows in plotting:
        rows = ''.join(rows).split()
        X.append(float(rows[0]))
        Y.append(float(rows[1]))
plt.tight_layout()
plt.axis([-2500,2500,-2500,2500])
plt.title('Распределение вихрей в сверхпроводнике, U0*0.03')
plt.scatter(X,Y)
```
