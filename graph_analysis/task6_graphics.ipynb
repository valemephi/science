```python
import pandas as pd
import numpy as np
import seaborn as sns
import csv
from matplotlib import pyplot as plt
import matplotlib.animation as animation
from matplotlib.animation import FuncAnimation
from IPython import display
```

```python
X1 = []
Y1 = []
X2 = []
Y2 = []
with open('res1.txt', 'r') as datafile:
    plotting = csv.reader(datafile)
    for rows in plotting:
        rows = ''.join(rows).split()
        if float(rows[2])<0.0:
          X1.append(float(rows[0]))
          Y1.append(float(rows[1]))
        else:
          X2.append(float(rows[0]))
          Y2.append(float(rows[1]))
plt.tight_layout()
plt.axis([-2500,2500,-2500,2500])
#plt.title('Распределение вихрей в сверхпроводнике')
plt.scatter(X1,Y1,c='blue')
plt.scatter(X2,Y2,c='blue')
plt.grid()
plt.xlabel('$x, нм$')
plt.ylabel('$y, нм$')
```

```python
X1=[]
xi1=[]
Y1=[]
yi1=[]
X2=[]
xi2=[]
Y2=[]
yi2=[]
with open('resanim.txt', 'r') as datafile:
    plotting = csv.reader(datafile)
    for rows in plotting:
      if ''.join(rows) != ' flag':
        rows = ''.join(rows).split()
        if float(rows[2])<0.0:
          xi1.append(float(rows[0]))
          yi1.append(float(rows[1]))
        else:
          xi2.append(float(rows[0]))
          yi2.append(float(rows[1]))
      else:
        X1.append(xi1)
        Y1.append(yi1)
        X2.append(xi2)
        Y2.append(yi2)
        xi1=[]
        yi1=[]
        xi2=[]
        yi2=[]

fig, ax = plt.subplots()
ax.set_xlim(-2500,2500)
ax.set_ylim(-2500,2500)
ims = []
frames=len(X1)
def animate(i):
  ax.clear()
  ax.scatter(X1[i], Y1[i], c='blue')
  ax.scatter(X2[i], Y2[i], c='blue')
  ax.set_xlim(-2500,2500)
  ax.set_ylim(-2500,2500)
  ax.grid()

ani = animation.FuncAnimation(fig, animate, frames=frames, interval=500)

ani.save('anim.gif', writer='pillow')
video = ani.to_html5_video()
html = display.HTML(video)
display.display(html)
```
