```python
def S_6(X2, Y2):
  N = len(X2)
  V = np.zeros((N,2))
  V[:,0] = np.array(X2)
  V[:,1] = np.array(Y2)
  R_mean = np.zeros(N)
  for i in range(N):
    V_i =np.zeros((N,2))
    vort_ = V[i,:]
    # print(vort_)
    V_i[:, 0] = V[:, 0]-vort_[0]
    V_i[:, 1] = V[:, 1]-vort_[1]
    # print(V_i[:5,:])
    R_i = [np.sqrt(V_i[k,0]**2+V_i[k,1]**2) for k in range(N)]
    Nei = sorted(R_i)[1:7]
    R_mean[i] = np.mean(Nei)
  r_max = np.mean(R_mean)
  S6= 0.
  for i in range(N):
    V_i =np.zeros((N,2))
    vort_ = V[i,:]
    V_i[:, 0] = V[:, 0]-vort_[0]
    V_i[:, 1] = V[:, 1]-vort_[1]

    Nei = np.array([list(u) for u in sorted((V_i), key=lambda x: x[0]**2+x[1]**2)[1:7]])
    R_i = np.linalg.norm(Nei, axis=1)
    Nei_i = Nei[np.where(R_i<r_max)[0], :]
    R_i = R_i[np.where(R_i<r_max)[0]]
    Z = len(Nei_i)
    theta = np.arctan(Nei_i[:,1]/Nei_i[:,0])
    S6 += 1/Z*sum(complex(np.cos(6*theta[i]),np.sin(6*theta[i])) for i in range(Z))
  return np.sqrt((S6/N).imag**2+(S6/N).real**2)
```
