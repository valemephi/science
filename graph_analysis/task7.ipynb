```python
import pandas as pd
import numpy as np
import seaborn as sns
import csv
from matplotlib import pyplot as plt
import matplotlib.animation as animation
from matplotlib.animation import FuncAnimation
from IPython import display
```

```python
X = []
Y = []
X_anti = []
Y_anti = []
flag = 0
with open('res1.txt', 'r') as datafile:
    plotting = csv.reader(datafile)
    for rows in plotting:
        rows = ''.join(rows).split()

        if len(rows)==1:
          flag =1
          continue
        if flag:
          X_anti.append(float(rows[0]))
          Y_anti.append(float(rows[1]))
        else:
          X.append(float(rows[0]))
          Y.append(float(rows[1]))
plt.tight_layout()
# plt.scatter(X, Y,c='red', lw=6)
plt.axis([-2500,2500,-2500,2500])
plt.xticks(list(range(-2500,3000,500)))
#plt.title('Распределение вихрей в сверхпроводнике')
plt.scatter(X,Y,c='red', lw=5)
plt.scatter(X_anti,Y_anti,c='green')
plt.grid()
plt.xlabel('$x, нм$')
plt.ylabel('$y, нм$')
```

```python
length = 2.*10**3
width = 2.*10**3
lambda0 = 180.0
T=10.
Tc=84.
lambda1 = lambda0*(1-(T/Tc)**3.3)**(-1./2.)
alpha = 1.
xi0 = 2
xi = xi0*(1-(T/Tc)**3.3)**(-1./2.)
phi0 = 2.06783460999999E-07
delta = 0.27
U0 = phi0**2/(8.*np.pi**2*lambda1**2*10.**(-14)) * delta*10.**(-7)


x = np.linspace(10**(-9), 5*xi*10**(-7), 10000)
Udef = -alpha*U0*(1/(x/(xi*10.**(-7))+1))*np.exp(-x*10**(7)/(2*xi))
mas = Udef*6.2*10**(11)
np.savetxt('Uptabpy.txt', mas)
plt.plot(x, Udef)
plt.plot(mas)
```

```python
xdefs = np.linspace(-length/2+0.45*lambda1, length/2-0.45*lambda1, 10)
ydefs = np.linspace(-width/2+0.45*lambda1, width/2-0.45*lambda1, 10)
a=np.zeros((100,2))
for i in range(10):
  for j in range(10):
    a[10*i+j, :] = (xdefs[i], ydefs[j])
X=a[:,0]
Y=a[:,1]
M=len(a)
b=[]
with open("defs2.txt", "w") as f:
    f.write(str(M))
    f.write('\n')
    for i in range(M):
      f.write(str(a[i,0])+' '+ str(a[i,1])+'\n')
      b.append(str(a[i,0])+' '+ str(a[i,1])+'\n')
print(b)

plt.scatter(X,Y)
```
