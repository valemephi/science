```python
import pandas as pd 
import numpy as np
import seaborn as sns
import csv
from matplotlib import pyplot as plt
f = open("heatmap_task1_data.txt")
a = []
for line in f:
    a.append(line.split())
f1 = open("heatmap_data.txt", 'w')
for i in a:
    f1.write(','.join(i))
    f1.write('\n')
f.close()
f1.close()
 
```

```python
X = []
Y = []
with open('res2.txt', 'r') as datafile:
    plotting = csv.reader(datafile)
    for rows in plotting:
        rows = ''.join(rows).split()
        X.append(float(rows[0]))
        Y.append(float(rows[1]))
plt.tight_layout()
plt.axis([-2500,2500,-2500,2500])
plt.title('Distribution of vortices in superconductor')
plt.scatter(X,Y)
```

```python
X = []
Y = []
with open('task1energy.txt', 'r') as datafile:
    plotting = csv.reader(datafile)
    for rows in plotting:
        rows = ''.join(rows).split()
        X.append(float(rows[0]))
        Y.append(float(rows[1]))
plt.ylabel('Energy, eV')
plt.xlabel('Iterations')
plt.title('Energy of the system')
plt.plot(X, Y)
```
